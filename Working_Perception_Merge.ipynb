```python
percept2_df = pd.read_csv('Perceptions_Other_England and Wales_2024Q4.csv')
```

```python
percept2_df['MeasurementVar'].unique()
```

```python
ref_data = pd.read_csv("reference-measurementvar-with-categories.csv")
```

```python
merge2_df = pd.merge(percept2_df, ref_data, how='left', left_on='MeasurementVar', right_on='MeasurementVar')
```

```python
merge2_df
```

```python
duplicates = merge2_df[merge2_df.duplicated()]
duplicates_count = duplicates.shape[0]
print(f"Number of duplicate rows: {duplicates_count}")
```

```python
merge2_df.columns
```

```python
merge2_df['MeasurementLevel_x'].unique()
```

```python
merge2_df = merge2_df.drop(columns=['MeasurementVar', 'MeasurementType_x', 'MeasurementLevel_x', 'MeasurementLevel_y', 'MeasurementType_y',
                                     'Sex', 'HouseholdType', 'Quarter', 'Geography'])
merge2_df.head()
```

The estimate is shown as a percentage

```python
merge2_df = merge2_df[(merge2_df['Characteristic'] == 'Yorkshire and The Humber')]
merge2_df.head()
```

```python
# Round estimate to 2 decimal places and show as percentage
# The value shown is already calculated as a percentage, so we just need to format it
merge2_df['Estimate'] = merge2_df['Estimate'].astype(float)  # Ensure it's float before rounding
# Round to 2 decimal places and convert to string with percentage sign
merge2_df['Estimate'] = merge2_df['Estimate'].round(2)  
merge2_df['Estimate'] = merge2_df['Estimate'].astype(str) + '%' 
```

```python
merge2_df
```

```python
merge2_df = merge2_df.drop(columns=['CharacteristicVar', 'StandardError', 'UnweightedCount'])
```

```python
# Save the final DataFrame to a CSV file
merge2_df.to_csv('Perceptions_Data_Yorkshire.csv', index=False) 
# Read the saved CSV file to verify
saved_df = pd.read_csv('Perceptions_Data_Yorkshire.csv')
```
