```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Crime data
crime_df = pd.read_csv("yorkshire_crime_2024.csv")

# Perception reference data
reference_df = pd.read_csv("reference-measurementvar.csv")
```

```python
label_to_category = {
    'Perceived problem with abandoned or burnt out vehicles': 'vehicle-crime',
    'Worried about burglary': 'burglary',
    'Worried about car crime': 'vehicle-crime',
    'Worried about fraud': 'other-crime',
    'Worried about violent crime': 'violent-crime',
    'Perceived problem with  people using or dealing drugs': 'drugs',
    'Perceived problem with people being drunk or rowdy': 'anti-social-behaviour',
    'Perceived high level of anti-social behaviour': 'anti-social-behaviour',
    'Perceived problem with noisy neighbours': 'anti-social-behaviour',
    'Perceived problem with teenagers hanging around': 'anti-social-behaviour',
    'Perceived problem with vandalism and graffiti': 'criminal-damage-arson',
    'Perceived problem with litter or rubbish': 'anti-social-behaviour',
}
reference_df["MatchedCategory"] = reference_df["MeasurementLabel"].map(label_to_category)
```

```python
# Load perception scores (if in another file)
perception_scores = pd.read_csv("perception_df.csv")

# Merge onto reference with categories
merged_ref = reference_df.merge(perception_scores, on="MeasurementVar", how="left")
```

```python
crime_summary = crime_df.groupby(["force", "category"]).size().reset_index(name="crime_count")
```

```python
# Merge on category
final_df = crime_summary.merge(reference_df[["MatchedCategory", "MeasurementVar"]], 
                                left_on="category", right_on="MatchedCategory", how="left")
```

```python
# % of total crime by force
crime_pct = final_df.groupby("force").apply(
    lambda x: x.assign(crime_pct=x["crime_count"] / x["crime_count"].sum() * 100)
).reset_index(drop=True)

# If perception data available: compare crime_pct vs perception
```

```python
crime_pct.to_csv("crime_vs_perception_by_force.csv", index=False)
```
